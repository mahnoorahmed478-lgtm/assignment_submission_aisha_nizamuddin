1) A company decided to give bonus of 5% to employee if his/her year of service is more than 5 years.
Ask user for their salary and year of service and print the net bonus amount.

```python
salary = int(input("Enter your salary: "))
years_of_service = int(input("Enter your years of service: "))

print("Input Values:")
print("Salary:", salary)
print("Years of Service:", years_of_service)

if years_of_service > 5:
    bonus = salary * 5 / 100
    print("\nOutput:")
    print("Net bonus amount:", bonus)
else:
    print("\nOutput:")
    print("No bonus. Years of service must be more than 5.")
```

11) A shop will give discount of 10% if the cost of purchased quantity is more than 1000.
Ask user for quantity
Suppose, one unit will cost 100.
Judge and print total cost for user.

```python
quantity = int(input("Enter the quantity purchased: "))
unit_price = 100

total_cost = quantity * unit_price

print("Input Value:")
print("Quantity:", quantity)
print("Unit Price:", unit_price)

print("\nOutput:")
if total_cost > 1000:
    discount = total_cost * 10 / 100
    total_cost_after_discount = total_cost - discount
    print("Discount: 10%")
    print("Total cost after discount:", total_cost_after_discount)
else:
    print("No discount")
    print("Total cost:", total_cost)
```

14)A student will not be allowed to sit in exam if his/her attendence is less than 75%.

Take following input from user

- Number of classes held

- Number of classes attended.

And print

- percentage of class attended

- Is student is allowed to sit in exam or not.


```python
classes_held = int(input("Enter total number of classes held: "))
classes_attended = int(input("Enter number of classes attended: "))

print("Input Values:")
print("Classes Held:", classes_held)
print("Classes Attended:", classes_attended)

attendance_percentage = (classes_attended / classes_held) * 100

print("\nOutput:")
print("Attendance Percentage:", attendance_percentage, "%")

if attendance_percentage >= 75:
    print("Student is allowed to sit in the exam")
else:
    print("Student is not allowed to sit in the exam")
```

16) Write a program to check if a year is leap year or not.

If a year is divisible by 4 then it is leap year but if the year is century year like 2000, 1900, 2100 then it must be divisible by 400.

```python
year = int(input("Enter a year: "))

print("Input Value:")
print("Year:", year)

print("\nOutput:")
if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "is a leap year")
        else:
            print(year, "is not a leap year")
    else:
        print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

13) Take input of age of 3 people by user and determine oldest and youngest among them.

```python
age1 = int(input("Enter age of first person: "))
age2 = int(input("Enter age of second person: "))
age3 = int(input("Enter age of third person: "))

print("Input Values:")
print("Age1:", age1)
print("Age2:", age2)
print("Age3:", age3)

print("\nOutput:")
oldest = max(age1, age2, age3)
youngest = min(age1, age2, age3)

print("Oldest age:", oldest)
print("Youngest age:", youngest)
```
