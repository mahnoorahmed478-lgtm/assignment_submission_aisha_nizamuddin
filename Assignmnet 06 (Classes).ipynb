1. Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
# ANSWER CODE 

# Created a class named Circle
class Circle:

    # Constructor to take radius input
    def __init__(self, radius):
        self.radius = radius

    # Method to calculate area
    def area(self):
        area = 3.14 * self.radius * self.radius
        return area

    # Method to calculate perimeter (circumference)
    def perimeter(self):
        perimeter = 2 * 3.14 * self.radius
        return perimeter


# Taking input from user
r = float(input("Enter the radius of the circle: "))

# Creating object of Circle class
c = Circle(r)

# Printing results
print("Area of Circle:", c.area())
print("Perimeter of Circle:", c.perimeter())
```

2. Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.


```python
# ANSWER of the Code

from datetime import date

# Create Person class
class Person:

    # Constructor
    def __init__(self, name, country, birth_year):
        self.name = name
        self.country = country
        self.birth_year = birth_year

    # Method to calculate age
    def calculate_age(self):
        current_year = date.today().year
        age = current_year - self.birth_year
        return age


# Taking input from user
name = input("Enter your name: ")
country = input("Enter your country: ")
birth_year = int(input("Enter your birth year: "))

# Creating object
p = Person(name, country, birth_year)

# Output
print("Name:", p.name)
print("Country:", p.country)
print("Age:", p.calculate_age())
```

3. Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python
# Add your code here

# Create Calculator class
class Calculator:

    # Addition
    def add(self, a, b):
        return a + b

    # Subtraction
    def subtract(self, a, b):
        return a - b

    # Multiplication
    def multiply(self, a, b):
        return a * b

    # Division
    def divide(self, a, b):
        return a / b


# Create object
c = Calculator()

# Taking input from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Output
print("INPUT 1 is: ",num1)
print("INPUT 2 is: ",num2)
print("Addition:", c.add(num1, num2))
print("Subtraction:", c.subtract(num1, num2))
print("Multiplication:", c.multiply(num1, num2))
print("Division:", c.divide(num1, num2))
```
