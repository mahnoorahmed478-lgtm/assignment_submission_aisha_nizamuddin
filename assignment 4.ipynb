## Question 1 

```python
# Take input from user
name = input("Enter student name: ")
age = int(input("Enter student age: "))
course = input("Enter student course: ")
city = input("Enter student city: ")

# Create dictionary
student = {
    "name": name,
    "age": age,
    "course": course,
    "city": city
}

# Output
print("\nStudent Details:")
print("Name:", student["name"])
print("Age:", student["age"])
print("Course:", student["course"])
print("City:", student["city"])
```

## Question 4

```python
numbers = [1,2,2,3,3,3,4,4,4,4]

frequency = {}

for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1

print(frequency)
```

## Question 5

```python
sentence = input("Enter a sentence: ")

words = sentence.split()
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("\nYou entered:", sentence)
print("Word Frequency:", word_count)
```

## Question 6

```python
marks = {"Math":80, "English":70, "Physics":90}

total = sum(marks.values())
average = total / len(marks)

print("Total Marks:", total)
print("Average Marks:", average)
```

## Question 7

```python
students = {
    "Ali":85,
    "Sara":92,
    "Ahmed":78,
    "Zara":95
}

highest = max(students, key=students.get)

print("Highest Scoring Student:", highest)
print("Marks:", students[highest])
```

## Question 9

```python
contacts = {}

# Add 3 contacts
for i in range(3):
    name = input("Enter contact name: ")
    phone = input("Enter phone number: ")
    contacts[name] = phone

# Search contact
search_name = input("Enter name to search: ")

print("\nYou searched for:", search_name)

if search_name in contacts:
    print("Phone Number:", contacts[search_name])
else:
    print("Contact not found")
```
