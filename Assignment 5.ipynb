## QUESTION 1: Working with Lists

```python
# Create a list called numbers with 8 integers [cite: 2]
numbers = [12, 7, 20, 15, 8, 3, 50, 22]

# Initialize a variable to store the sum
total_sum = 0

print("All numbers in the list:")
for num in numbers:
    # Print the current number [cite: 4]
    print(num)
    
    # Add the current number to the sum [cite: 5]
    total_sum = total_sum + num
```

## QUESTION 2: Dictionary Basics

```python
# Q2: Student Dictionary
student = {
    "name": "Alex",
    "age": 20,
    "course": "Computer Science",
    "marks": [85, 42, 78]  # List inside dictionary
}

# Calculate average using a loop
mark_sum = 0
for m in student["marks"]:
    mark_sum += m

average = mark_sum / len(student["marks"])
print(f"Average: {average}")

# Passing logic (average >= 50)
if average >= 50:
    print("Result: Passed")
else:
    print("Result: Failed")
```

## QUESTION 3: List of Dictionaries

```python
# Create a list of employee dictionaries [cite: 25, 26]
employees = [
    {"name": "Alice", "department": "HR", "salary": 50000},
    {"name": "Bob", "department": "IT", "salary": 75000},
    {"name": "Charlie", "department": "Design", "salary": 60000}
]

total_expense = 0
highest_salary = 0
richest_employee = ""

print("Employee Details:")
for emp in employees:
    # Print details for each [cite: 31]
    print(f"Name: {emp['name']}, Dept: {emp['department']}, Salary: {emp['salary']}")
    
    # Calculate total expense [cite: 33]
    total_expense += emp['salary']
    
    # Logic to find highest salary [cite: 32, 36]
    if emp['salary'] > highest_salary:
        highest_salary = emp['salary']
        richest_employee = emp['name']

print(f"\nTotal Salary Expense: {total_expense}")
print(f"Highest Salary: {highest_salary} (Employee: {richest_employee})")
```

## QUESTION 4: While Loop with User Input

```python
user_numbers = []

# Loop until user enters -1 [cite: 41, 47]
while True:
    val = int(input("Enter a number (or -1 to stop): "))
    if val == -1:
        break
    user_numbers.append(val) # Store in list [cite: 40, 48]

if user_numbers:
    print(f"Complete List: {user_numbers}")
    print(f"Largest Number: {max(user_numbers)}") # [cite: 44]
    print(f"Smallest Number: {min(user_numbers)}") # [cite: 45]
```

## QUESTION 5: Word Frequency Counter

```python
sentence = input("Enter a sentence: ")
words = sentence.split() # Convert to list of words [cite: 52, 58]
frequency = {}

# Use loop to process data [cite: 60]
for word in words:
    word = word.lower() # Normalize to lowercase
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print("Word Frequency:")
for word, count in frequency.items():
    print(f"{word}: {count}") # [cite: 54]
```

## QUESTION 6: Nested Loops – Multiplication Table

```python
# Nested loops for rows and columns [cite: 62, 66]
for i in range(1, 6): # Outer loop (rows)
    for j in range(1, 6): # Inner loop (columns)
        # Calculate and format with a space [cite: 64, 67]
        print(i * j, end="\t") 
    print() # New line after each row
```
