## 1. Print alternate elements of a list

```python
lst = list(map(int, input("Enter list elements: ").split()))

print("Alternate elements:")
for i in range(0, len(lst), 2):
    print(lst[i])
```

## 2. Reverse a list without using reverse()

```python
lst = list(map(int, input("Enter list elements: ").split()))
rev = []

for i in range(len(lst)-1, -1, -1):
    rev.append(lst[i])

print("Reversed list:", rev)
```

## 3. Find largest number without using max()

```python
lst = list(map(int, input("Enter list elements: ").split()))

largest = lst[0]
for num in lst:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

## 4. Rotate list (last element to first)

```python
lst = list(map(int, input("Enter list elements: ").split()))

rotated = [lst[-1]] + lst[:-1]
print("Rotated list:", rotated)
```

## 5. Delete a given word from a string

```python
sentence = input("Enter a string: ")
word = input("Enter word to delete: ")

new_sentence = sentence.replace(word, "")
print("Updated string:", new_sentence)
```

## 6. Convert date format (mm/dd/yyyy â†’ Month day, year)

```python
date = input("Enter date (mm/dd/yyyy): ")

months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

m, d, y = date.split("/")
print(months[int(m)-1], d + ",", y)
```

## 7. Capitalize first letter of each word (using function)

```python
def capitalize_words(s):
    words = s.split()
    result = ""

    for w in words:
        result += w.capitalize() + " "

    return result.strip()

sentence = input("Enter a sentence: ")
print(capitalize_words(sentence))
```

## 8. Sum of each row of a matrix

## 9. Add two matrices

```python
r = int(input("Enter rows: "))
c = int(input("Enter columns: "))

print("Enter first matrix:")
A = [list(map(int, input().split())) for _ in range(r)]

print("Enter second matrix:")
B = [list(map(int, input().split())) for _ in range(r)]

result = []
for i in range(r):
    row = []
    for j in range(c):
        row.append(A[i][j] + B[i][j])
    result.append(row)

print("Sum of matrices:")
for row in result:
    print(row)
```

## 10. Multiply two matrices

```python
r1 = int(input("Rows of first matrix: "))
c1 = int(input("Columns of first matrix: "))
r2 = int(input("Rows of second matrix: "))
c2 = int(input("Columns of second matrix: "))

if c1 != r2:
    print("Matrix multiplication not possible")
else:
    print("Enter first matrix:")
    A = [list(map(int, input().split())) for _ in range(r1)]

    print("Enter second matrix:")
    B = [list(map(int, input().split())) for _ in range(r2)]

    result = [[0 for _ in range(c2)] for _ in range(r1)]

    for i in range(r1):
        for j in range(c2):
            for k in range(c1):
                result[i][j] += A[i][k] * B[k][j]

    print("Product of matrices:")
    for row in result:
        print(row)
```
